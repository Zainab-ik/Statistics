```python
import math
import random
import matplotlib.pyplot as plt  
```

```python
def GenerateRandomData(sz):
    import numpy as np
    return [x for x in np.random.binomial(n=100, p=0.5, size=sz)] 
```

```python
data=GenerateRandomData(10000) 
```

```python
def data_range(x):
    return max(x) - min(x) 
```

```python
print(data_range(data)) 
```

```python
def quantile(p,x):
    sorted_x=sorted(x)
    index=int(p*len(x))
    return sorted_x[index] 
```

```python
def iqr(x):
    return quantile(0.75,x)-quantile(0.25,x)

    
```

```python
print(iqr(data)) 
```

```python
def mean(x):
    return sum(x_i for x_i in x)/len(x)
```

```python
def simplevariance(x):
    u=mean(x)
    return sum((x_i-u)**2 for x_i in x)/len(x)-1 
```

```python
print(simplevariance(data)) 
```

```python
def standardDeviation(x):
    return math.sqrt(simplevariance(x)) 
```

```python
print(standardDeviation(data)) 
```

```python
def NormalDistributionFunction(x,u,s):
    sqr_2pi = math.sqrt(2*math.pi)
    return math.exp(-((x-u)**2)/(2*s*s))/(s*sqr_2pi)
```

```python
def PlotNormalDistribution(u,s):
    data_x = [x/10 for x in range((u-4*s)*10,(u+4*s)*10)]
    plt.plot(data_x, [NormalDistributionFunction(x,u,s) for x in data_x],label="Normal Distribution 1" )
    plt.title("Normal Distribution Graph")
    plt.legend(loc='best')   
```

```python
PlotNormalDistribution(0,2)
plt.show()  
```

```python
PlotNormalDistribution(0,1)
PlotNormalDistribution(-2,2)
plt.show()   
```

```python
PlotNormalDistribution(0,10)
PlotNormalDistribution(-2,20)
PlotNormalDistribution(3,5)
plt.show()  
```
